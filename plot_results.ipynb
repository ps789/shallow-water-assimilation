```python
import numpy as np
cnn_full = np.load("results/cnn_full_rmse.npy", allow_pickle = True)
cnn_sample = np.load("results/cnn_sample_rmse.npy", allow_pickle = True)
pca_full = np.load("results/pca_full_rmse.npy", allow_pickle = True)
cnn_sample_kld = np.load("results/cnn_sample_pluskld_rmse.npy", allow_pickle = True)
```

```python
import matplotlib.pyplot as plt
plt.plot(cnn_full, "b-", label = "CNN Full")
plt.plot(cnn_sample, "b--", label = "CNN Sparse")
plt.plot(cnn_sample_kld, "g--", label = "CNN Sparse with KLD")
plt.plot(pca_full, "r-", label = "PCA Full")
plt.title("Comparison of Reduced-Order Assimilation with PCA and also Model Trained with KLD")
plt.legend()
plt.savefig("./results/main_comparison.png")
plt.show()
```
